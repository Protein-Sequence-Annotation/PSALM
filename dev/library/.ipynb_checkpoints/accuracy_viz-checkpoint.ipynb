```python
from pathlib import Path
from Bio import SeqIO
# import pandas as pd
import pickle
import polars as pl
```

## Retrieve sequences by split and filter out BLAST hits

```python
pid_splits = [20, 40, 60, 80, 100]
num_shards = 50

split_seqs = {}

for pid in pid_splits:
    folder = Path(f'../data/test_fasta_{pid}')
    if pid != 100:
        df = pd.read_csv(f"{folder}/filtered_entries.txt",sep="\t", header=None)
        exclude_seqs = list(df[0])
    else:
        exclude_seqs = []

    split_seqs[pid] = []
    for i in range(1, num_shards+1):
        for record in SeqIO.parse(folder / f'split_{i}_test_ids_full.fasta', 'fasta'):
            if record.id not in exclude_seqs:
                split_seqs[pid].append(record.id)
```

## Filter out possible ESM-2 Sequences

```python
import pickle
with open(f'../data/non_uniref.pkl', 'rb') as f:
    non_uniref = pickle.load(f)
```

```python
len(split_seqs[20])
```

```python
non_uniref_seqs = list(non_uniref['unique'])
from tqdm import tqdm

def non_uniref_split(seq_list):
    return [x for x in seq_list if x in non_uniref_seqs]

pid_splits = [20, 40, 60, 80, 100]

for pid in tqdm(pid_splits):
    split_seqs[pid] = non_uniref_split(split_seqs[pid])
```

### Non-clan filtering fxns

```python
with open('../data/benchmarking/old_test_pids.pkl', 'rb') as f:
    pid_dict = pickle.load(f)
with open('../data/maps.pkl', 'rb') as f:
    maps = pickle.load(f)
def nc(pid):
    result_nc = []
    for seq in split_seqs[pid]:
        if "NC0001" in [maps["fam_clan"][x] for x in pid_dict[seq].keys()]:
            result_nc.append(seq)
    return(result_nc)
def not_nc(pid):
    result_nc = []
    for seq in split_seqs[pid]:
        if "NC0001" not in [maps["fam_clan"][x] for x in pid_dict[seq].keys()]:
            result_nc.append(seq)
    return(result_nc)
```

```python
import sys
import numpy as np
import hmmscan_utils as hu
import classifiers as cf
import ml_utils as mu
import visualizations as vz
import torch
import pickle
from pathlib import Path

with open('../data/shard_seqs2.pkl', 'rb') as f:
    shard_seqs = pickle.load(f)

with open(f'../data/maps.pkl', 'rb') as g:
        maps = pickle.load(g)

fam_map = list(maps['fam_idx'].keys())
clan_map = list(maps['clan_idx'].keys())

device = 'cuda' if torch.cuda.is_available() else 'cpu'

length_limit = 4096 # Covers 99.75% sequences
model_name =  'esm2_t33_650M_UR50D'
num_shards = 50 
data_utils = mu.DataUtils('../data', num_shards, model_name, length_limit, 'test', device,"_OLD")

fam_classifier = cf.FamModelSimple(data_utils.embedding_dim, data_utils.maps, device).to(device)
fam_classifier_path = Path('../data/results/fam_finetune/epoch_3.pth')
fam_classifier.load_state_dict(torch.load(fam_classifier_path))

clan_classifier = cf.ClanLSTM(data_utils.embedding_dim,data_utils.clan_count).to(device)
clan_classifier_path = Path('../data/results/clan_finetune/epoch_5.pth')
clan_classifier.load_state_dict(torch.load(clan_classifier_path))

fam_classifier.eval()
clan_classifier.eval()
```

```python
with open(f'../data/split_seqs.pkl' , 'rb') as f:
    split_seqs = pickle.load(f)
    
current_set = split_seqs[20]
```

```python
with torch.inference_mode():
    
    seq_id = 'I4D8X5.1'
    # seq_id = np.random.choice(current_set)
    shard = shard_seqs[seq_id]
    # shard = np.random.randint(1,51)
    # shard = 17

    hmm_dict = data_utils.parse_shard(shard)
    pred_dict = hu.parse_hmmscan_results(f"../data/test_hmmer_preds/split_{shard}_test_ids_full.fasta_scan.txt", pred=True)
    # keys = list(hmm_dict.keys())
    # key_idx = np.random.randint(0,len(keys))
    # seq_id = keys[key_idx]
    # seq_id = 'R6B7I5.1'

    dataset = data_utils.get_dataset(shard)
    dataset = data_utils.filter_batches(dataset, [seq_id])
    data_loader = data_utils.get_dataloader(dataset)

    label, seq, token = next(iter(data_loader))
    token = token.to(device)
    embedding = data_utils.get_embedding(token)
    stop_index = min(len(seq[0]), data_utils.length_limit)
    fam_vector, clan_vector = hu.generate_domain_position_list2(hmm_dict, seq_id, data_utils.maps) # Target
    hmm_preds, _ = hu.generate_domain_position_list2(pred_dict, seq_id, data_utils.maps) # HMMER* pred
    hmm_preds = torch.tensor(hmm_preds[:stop_index,:])
                    
    
    fam_vector = torch.tensor(fam_vector[:stop_index,:]).to(device) # clip the clan_vector to the truncated sequence length
    clan_vector = torch.tensor(clan_vector[:stop_index,:]).to(device)

    clan_preds = clan_classifier(embedding["representations"][data_utils.last_layer][0,1:stop_index+1,:])
    clan_preds = torch.nn.functional.softmax(clan_preds, dim=1)

    _, raw_preds = fam_classifier(embedding["representations"][data_utils.last_layer][0,1:stop_index+1,:], clan_preds)
    # fam_preds = torch.nn.functional.softmax(fam_preds, dim=1)
    
    ##CLAN_CHUNK_FAM
    clan_fam_weights = data_utils.maps['clan_family_matrix'].to(device)
    # print(torch.nonzero(clan_fam_weights[-1]).squeeze())
    for i in range(clan_fam_weights.shape[0]): #shape is cxf (ignore IDR bc 1:1 map)
        indices = torch.nonzero(clan_fam_weights[i]).squeeze() #indices for the softmax
        if i == 656:
            raw_preds[:,indices] = 1
        else:
            raw_preds[:,indices] = torch.softmax(raw_preds[:,indices],dim=1)

    #########

    #########
    # Multiply by clan
    # expand clan preds
    clan_preds_f = torch.matmul(clan_preds,clan_fam_weights)
    # element wise mul of preds and clan
    fam_preds = raw_preds * clan_preds_f
    #########
    # preds = preds.cpu()
    
    # Fam Results
    results_fam = {}
    results_fam[seq_id] = {}
    top_two_vals, top_two_indices = torch.topk(fam_preds, k=2, dim=1)

    results_fam[seq_id]['fam_vals'] = top_two_vals.cpu().numpy()
    results_fam[seq_id]['fam_idx'] = top_two_indices.cpu().numpy()
    results_fam[seq_id]['fam_true'] = torch.argmax(fam_vector, dim=1).cpu().numpy()
    results_fam[seq_id]['fam_true_vals'] = fam_vector.max(dim=1)[1].cpu().numpy()
    results_fam['label'] = seq_id
    results_fam['shard'] = shard
    
    # Clan Results
    results_clan = {}
    results_clan[seq_id] = {}
    top_two_vals, top_two_indices = torch.topk(clan_preds, k=2, dim=1)

    results_clan[seq_id]['clan_vals'] = top_two_vals.cpu().numpy()
    results_clan[seq_id]['clan_idx'] = top_two_indices.cpu().numpy()
    results_clan[seq_id]['clan_true'] = torch.argmax(clan_vector, dim=1).cpu().numpy()
    results_clan[seq_id]['clan_true_vals'] = clan_vector.max(dim=1)[1].cpu().numpy()
    results_clan['label'] = seq_id
    results_clan['shard'] = shard

    # HMMER* Results
    results_hmm = {}
    results_hmm[seq_id] = {}
    top_two_vals, top_two_indices = torch.topk(hmm_preds, k=2, dim=1)

    results_hmm[seq_id]['hmm_vals'] = top_two_vals.cpu().numpy()
    results_hmm[seq_id]['hmm_idx'] = top_two_indices.cpu().numpy()
    results_hmm[seq_id]['hmm_true'] = torch.argmax(fam_vector, dim=1).cpu().numpy()
    results_hmm[seq_id]['hmm_true_vals'] = fam_vector.max(dim=1)[1].cpu().numpy()
    results_hmm['label'] = seq_id
    results_hmm['shard'] = shard

print(shard, seq_id)
vz.viewAll(shard, seq_id, results_clan, results_fam, results_hmm, fam_map, clan_map, 'old_test_pids.pkl')
```

```python
# Assume negs is your dictionary
# Generate a list of thresholds
thresholds = np.linspace(0, 1, 1000)

# Initialize lists to store false positive rates
fp_rates = []

for threshold in thresholds:
    fam_negs = []
    for seq in negs.keys():
        fam_idx = negs[seq]["fam_idx"]
        clan_idx = negs[seq]["clan_idx"]
        thresh = negs[seq]["fam_preds"] > threshold
        score = ((fam_idx != 19632) & thresh).sum()
        score = score / len(fam_idx)
        fam_negs.append(score)
    # Calculate false positive rate and append to list
    fp_rate = np.mean(fam_negs)
    fp_rates.append(fp_rate)


# Convert the false positive rates to a numpy array
fp_rates = np.array(fp_rates)

# Find the index of the false positive rate closest to 1%
index = np.abs(fp_rates - 0.01).argmin()

# Get the corresponding threshold
threshold = thresholds[index]

# Create plot
plt.plot(fp_rates, thresholds, 'k--')  # 'k--' specifies a black dashed line
plt.axhline(threshold, color='r', linestyle='--')  # red dashed line
plt.axvline(0.01, color='r', linestyle='--')  # red dashed line

# Set x and y ticks at regular intervals
xticks = np.arange(0, np.max(fp_rates), 0.1)
yticks = np.arange(0, 1.2, 0.2)

# Add specific ticks
xticks = np.append(xticks, 0.01)
yticks = np.append(yticks, threshold)

# Remove duplicates and sort
xticks = np.unique(xticks)
yticks = np.unique(yticks)

plt.xticks(xticks)
plt.yticks(yticks)

plt.xlabel('False Positive Rate')
plt.ylabel('Residue Assignment Probability')
plt.title('Assignment Probability vs. FPR')
plt.show()
```

```python
def false_positive_rate(preds,len_thresh=4096, prob_thresh=0, norm=True, nc_only=False):
    fam_negs = []
    for seq in preds.keys():
        if len(preds[seq]["fam_idx"]) > len_thresh:
            continue
        fam_idx = preds[seq]["fam_idx"]
        clan_idx = preds[seq]["clan_idx"]
        thresh_idx = preds[seq]["fam_preds"]>prob_thresh
        if nc_only:
            score = ((fam_idx != 19632) & thresh_idx & (clan_idx != 655)).sum()
        else:
            score = ((fam_idx != 19632) & thresh_idx).sum()
        if norm:
            score = score/len(fam_idx)
        fam_negs.append(score)
    return fam_negs
```
